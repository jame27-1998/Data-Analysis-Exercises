```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('Udemy Courses.csv')
```

```python
data.head()
```

```python
#Q. 1) What are the different subjects that Udemy offers courses for?
data['subject'].unique()
```

```python
#Q. 2) Which subject has maximum number of courses?
data.subject.value_counts()
```

```python
#Q. 3) Show all the courses that are free.
data[data.price == 'Free']
```

```python
#Q. 4) Show all the courses that are paid.
data[data.is_paid == True]
```

```python
#Q. 5) Which are top selling courses?
data.sort_values('num_subscribers', ascending=False)
```

```python
#Q. 6) Which are least selling courses?
data.sort_values('num_subscribers', ascending=True)
```

```python
#Q. 7) Show all graphic design courses that are priced below 100.
data.price.unique()
```

```python
data['price'].replace(to_replace="Free", value= 0, inplace = True)
#Replace one word in a column with another in a particular column.
```

```python
data['price'] = pd.to_numeric(data.price)
#Change format of a column to numeric.
```

```python
data[(data.subject == 'Graphic Design') & (data.price > 100)]
```

```python
#Q. 8) Show all the courses related to Python.
# Filter dataset with a column containing certain strings.
data[data.course_title.str.contains("Python")]
```

```python
#Q. 9) What are courses published in year 2015?
#First way by keeping the Timestamp column as an object and filtering string containing "2015".
data[data.published_timestamp.str.contains("2015")]
```

```python
#Second way by making a year column from Timestamp and filtering int equaling "2015".
#First we will change the format to date time.
data['published_timestamp'] = pd.to_datetime(data.published_timestamp)
```

```python
#Now make a yeare column from it.
data['Year'] = data.published_timestamp.dt.year
```

```python
data[data.Year == 2015]
```

```python
#Q. 10) What are max subscribers for each level of course?
data.groupby('level')['num_subscribers'].max()
```

```python
data.groupby('level').max()
```
